```python
import numpy as np

def baum_welch(num_iterations, obs_seq_y, num_states, initial, transition_p, emission_q):
    len_obs_seq = len(obs_seq_y)

    for iteration in range(num_iterations):
        # E-step: Forward and Backward passes
        alpha = forward_algorithm(initial, transition_p, emission_q, obs_seq_y)
        beta = backward_algorithm(transition_p, emission_q, obs_seq_y)

        # M-step: Update parameters
        gamma = alpha * beta
        xi = np.zeros((num_states, num_states, len_obs_seq - 1))

        for n in range(len_obs_seq - 1):
            obs_prob = emission_q[:, obs_seq_y[n+1]]
            for i in range(num_states):
                for j in range(num_states):
                    xi[i, j, n] = alpha[i, n] * transition_p[i, j] * obs_prob[j] * beta[j, n+1]

        # Normalize gamma and xi
        gamma /= np.sum(gamma, axis=0, keepdims=True)
        xi /= np.sum(xi, axis=(0, 1), keepdims=True)

        # Update parameters
        initial = gamma[:, 0]
        transition_p = np.sum(xi, axis=2) / np.sum(gamma[:, :-1], axis=1, keepdims=True)
        for k in range(emission_q.shape[1]):
            emission_q[:, k] = np.sum(gamma[:, obs_seq_y == k], axis=1) / np.sum(gamma, axis=1)

    return initial, transition_p, emission_q

def forward_algorithm(initial, transition_p, emission_q, obs_seq_y):
    num_states = initial.shape[0]
    len_obs_seq = len(obs_seq_y)
    alpha = np.zeros((num_states, len_obs_seq))

    alpha[:, 0] = initial * emission_q[:, obs_seq_y[0]]

    for n in range(1, len_obs_seq):
        for s_n in range(num_states):
            for s_n_1 in range(num_states):
                alpha[s_n, n] += alpha[s_n_1, n-1] * transition_p[s_n_1, s_n] * emission_q[s_n, obs_seq_y[n]]

        norm1 = np.sum(alpha[:, n])
        alpha[:, n] = alpha[:, n] / norm1
    
   # print("aaaaaaaaaaaaaaaaa",alpha)
    return alpha

def backward_algorithm(transition_p, emission_q, obs_seq_y):
    num_states = transition_p.shape[0]
    len_obs_seq = len(obs_seq_y)
    beta = np.zeros((num_states, len_obs_seq))
    beta[:, len_obs_seq-1] = 1

    for n in range(len_obs_seq-2, -1, -1):
        for s_n in range(num_states):
            for s_n_1 in range(num_states):
                beta[s_n, n] += beta[s_n_1, n+1] * transition_p[s_n, s_n_1] * emission_q[s_n_1, obs_seq_y[n+1]]

        norm1 = np.sum(beta[:, n])
        beta[:, n] = beta[:, n] / norm1
    #print("bbbbbbbbbbbbbbbbbb",beta)
    return beta

# Example Usage
num_states = 2
num_iterations = 3

# Assume the values for Initial, Transition_p, Emission_q, and Obs_seq_y are already defined

# Perform Baum-Welch algorithm
new_initial, new_transition_p, new_emission_q = baum_welch(num_iterations, Obs_seq_y, num_states, Initial, Transition_p, Emission_q)

# Display the updated parameters
print("Updated Initial:")
print(new_initial)
print("\nUpdated Transition_p:")
print(new_transition_p)
print("\nUpdated Emission_q:")
print(new_emission_q)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# ... (Previous code remains unchanged)

def forward_algorithm(initial, transition_p, emission_q, obs_seq_y):
    num_states = initial.shape[0]
    len_obs_seq = len(obs_seq_y)
    alpha = np.zeros((num_states, len_obs_seq))
    norm2 = []
    alpha[:, 0] = initial * emission_q[:, obs_seq_y[0]]

    for n in range(1, len_obs_seq):
        for s_n in range(num_states):
            for s_n_1 in range(num_states):
                alpha[s_n, n] += alpha[s_n_1, n-1] * transition_p[s_n_1, s_n] * emission_q[s_n, obs_seq_y[n]]

        norm1 = np.sum(alpha[:, n])
        norm2.append(norm1)
        
        alpha[:, n] = alpha[:, n] / norm1
    
    #print("aaaaaaaaaaaaaaaaa",alpha)
    return alpha,norm2

def baum_welch_with_likelihood(num_iterations, obs_seq_y, num_states, initial, transition_p, emission_q):
    len_obs_seq = len(obs_seq_y)
    likelihood_values = []

    for iteration in range(num_iterations):
        # E-step: Forward and Backward passes
        alpha,norm2 = forward_algorithm(initial, transition_p, emission_q, obs_seq_y)
        #print("--------------------",norm2)
        beta = backward_algorithm(transition_p, emission_q, obs_seq_y)

        # Calculate likelihood for the current iteration
        #likelihood = np.prod(norm2)
        sum_norm_log = np.log(norm2)
        sum_of_logs = np.sum(sum_norm_log)
        #print("--sum of logs ---",sum_of_logs)
        likelihood = sum_of_logs
        likelihood_values.append(likelihood)

        # M-step: Update parameters
        gamma = alpha * beta
        xi = np.zeros((num_states, num_states, len_obs_seq - 1))

        for n in range(len_obs_seq - 1):
            obs_prob = emission_q[:, obs_seq_y[n+1]]
            for i in range(num_states):
                for j in range(num_states):
                    xi[i, j, n] = alpha[i, n] * transition_p[i, j] * obs_prob[j] * beta[j, n+1]

        # Normalize gamma and xi
        gamma /= np.sum(gamma, axis=0, keepdims=True)
        xi /= np.sum(xi, axis=(0, 1), keepdims=True)

        # Update parameters
        initial = gamma[:, 0]
        transition_p = np.sum(xi, axis=2) / np.sum(gamma[:, :-1], axis=1, keepdims=True)
        for k in range(emission_q.shape[1]):
            emission_q[:, k] = np.sum(gamma[:, obs_seq_y == k], axis=1) / np.sum(gamma, axis=1)

    return initial, transition_p, emission_q, likelihood_values

# Example Usage
num_states = 2
num_iterations = 100

# Perform Baum-Welch algorithm with likelihood tracking
new_initial, new_transition_p, new_emission_q, likelihood_values = baum_welch_with_likelihood(
    num_iterations, Obs_seq_y, num_states, Initial, Transition_p, Emission_q
)

# Display the updated parameters
print("Updated Initial:")
print(new_initial)
print("\nUpdated Transition_p:")
print(new_transition_p)
print("\nUpdated Emission_q:")
print(new_emission_q)

# Plot the likelihood values at each iteration
plt.plot(range(1, num_iterations + 1), likelihood_values)
plt.xlabel('Iteration')
plt.ylabel('Likelihood')
plt.title('Likelihood vs. Iteration')
plt.show()
```
